```python
# plan
# code: 11 different codes 
# pollutant: types of pollutant, 15 different types
# date: 24 time stamps per day, need to take an average for the pollutant type for these values to get a daily average for each reading per code

# remove NA values from 'value'
# create new column with date, no time called "new_date"
# collapse on 'code', 'new_data', and 'pollutant' to get average of value (i.e. group_by code, new_date, and pollutant (take mean of value in the process) )
```

```python
# gain insight into data_2020_2021
data_2020_2021.info
```

```python
# keep rows only where 'value' is not NA
data_2020_2021 = data_2020_2021[data_2020_2021['value'].notna()] 
```

```python
# gain insight into data_2020_2021 without NA values for 'value'
data_2020_2021.info
```

```python
# create new column, 'date_notime'
data_2020_2021['new_date'] = data_2020_2021['date'].str[0:10]
```

```python
# gain insight into data_2020_2021
data_2020_2021.head()
```

```python
# make new variable, data_grouped, by using group_by on data_2020_2021 to group by  'code', 'new_date', and 'pollutant' (take mean of value in the process)
data_grouped = data_2020_2021.groupby(['code','latitude', 'longitude', 'new_date', 'pollutant'], as_index=False)['value'].mean()
```

```python
# gain insight into data_grouped
data_grouped.info
```

```python
# gain insight into data_grouped
data_grouped.head(25)
```

```python
# recap
# now basically we have the average 'value' reading for each 'pollutant' on each 'new_date' for each metre (i.e 'code') across the GLA

# plan
# step 1
# we can now take an overall London 'value' average for each 'pollutant' by grouping by 'new_date' and 'pollutant', taking the mean of 'value', and ignoring 'code', 'latitude', and 'longitude'
# step 2
# plot each 'value' of each 'pollutant' type seperately over time (i.e. over 'new_date')
```

```python
# make new variable, gla_total, by using group_by on data_2020_2021 to group by 'new_date', and 'pollutant', taking the mean of 'value', and ignoring 'code', 'latitude', and 'longitude'
gla_total = data_2020_2021.groupby(['new_date', 'pollutant'], as_index=False)['value'].mean()
```

```python
# gain insight into gla_total
gla_total.head()
```

```python
pollutant_list = list(np.unique(gla_total['pollutant']))
pollutant_list
```

```python
# convert 'date_new' to date-time object
gla_total['new_date'] = pd.to_datetime(gla_total['new_date'])
```

```python
# remove weekends
gla_total = gla_total[gla_total['new_date'].dt.weekday < 5]
```

```python
gla_total.head()
```

```python
# plot each 'value' of each 'pollutant' type seperately over time (i.e. over 'new_date')
# air_temp
# create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# add x-axis and y-axis
ax.plot(gla_total['new_date'][gla_total['pollutant']=='air_temp'],
        gla_total['value'][gla_total['pollutant']=='air_temp'],
        color='purple')

# set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Average Air Temperature (Celcius)",
       title="Average Air Temperature in the Greater London Area: 2020-2021")

# rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```

```python
for metric in list(gla_total_long.columns.values):
    gla_total_long[f'SMA_3_{metric}'] = gla_total_long[metric].rolling(window=3).mean()
    gla_total_long[f'SMA_5_{metric}'] = gla_total_long[metric].rolling(window=5).mean()
```
